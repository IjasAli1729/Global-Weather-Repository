### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

SCALING

```python
data_copy = data.copy()
data_copy.isna().sum()
```

```python
data_copy1 = data1.copy()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_copy = pd.DataFrame(scaler.fit_transform(data_copy),columns =data.columns)
data_copy
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(data_copy,data_copy1, test_size=0.2, random_state= 42)
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(xtrain,ytrain)
knn.score(xtest,ytest)
```

ACCURACY RATE TABLE

```python
accuracy_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    score = knn.score(xtest,ytest)
    accuracy_rate.append(score)
```

```python
accuracy_rate
```

```python
most_accuracy = accuracy_rate.index(max(accuracy_rate)) + 1
most_accuracy
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
plt.plot(range(1,40),accuracy_rate,color = 'blue',marker = 'o', linestyle = ':', markerfacecolor = 'red', markersize = 8)
plt.title('Accuracy vs K vlaue')
plt.xlabel('K value')
plt.ylabel('Accuracy')
```

ERROR RATE TABLE

```python
import numpy as np
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    predict = knn.predict(xtest)
    error_rate.append(np.mean(predict != ytest))
```

```python
error_rate
```

```python
min_error = error_rate.index(min(error_rate))+ 1
min_error
```

```python
plt.figure(figsize=(12,7))
plt.plot(range(1,40),error_rate,color = 'red', linestyle = '--', marker = 'o', markerfacecolor = 'blue', markersize = 8)
plt.title('Error vs K value')
plt.xlabel("K value")
plt.ylabel('Error rate')
```
